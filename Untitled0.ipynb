Making the ideated model that consists convolutional and then LSTM.
For convolutional model, using RESNET50 and freezing its all layers and making fully connected last layer, which will be input to the LSTM

First importing libraries neccesary for RESNET50

```
from keras.applications.resnet50 import ResNet50
from keras.models import Model
import keras
```

COde for model instantiation

```
IMG_HEIGHT=IMG_WIDTH=300


restnet = ResNet50(include_top=False, weights='imagenet', input_shape=(IMG_HEIGHT,IMG_WIDTH,3))


output = restnet.layers[-1].output
output = keras.layers.Flatten()(output)
print(output)

restnet = Model(inputs=restnet.input, outputs=output)

for layer in restnet.layers:
    layer.trainable = False


restnet.summary()

```

NOW importing libraries for next sequence LSTM network

```
import numpy as np
#np.random.seed(0)
from keras.models import Model, Sequential
from keras.layers import Dense, Input, Dropout, LSTM, Activation
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
from keras.initializers import glorot_uniform
#np.random.seed(1)
```

```
model = Sequential()

model.add(restnet)

model.add(LSTM(128,input_shape = () return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(units = 128, return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])



```

```

input_shape = (224, 224, 3)

Input_img = Input(shape=input_shape, dtype='int32')

#flat1 = Flatten()(restnet.layers[-1].output)
class1 = Dense(1024, activation='relu')(restnet.layers[-1].output)

# The returned output should be a batch of sequences.
X = LSTM(units = 128, return_sequences=True)(class1)

# Add dropout with a probability of 0.5
X = Dropout(rate = 0.5)(X)

# Propagate X trough another LSTM layer with 128-dimensional hidden state
# The returned output should be a single hidden state, not a batch of sequences.
X = LSTM(units = 128, return_sequences=False)(X)

# Add dropout with a probability of 0.5
X = Dropout(rate= 0.5)(X)

# Propagate X through a Dense layer with 5 units
X = Dense(units=3)(X)

# Add a softmax activation
X = Activation('softmax')(X)

# Create Model instance which converts sentence_indices into X.
model = Model(inputs=restnet.input, outputs=X)

model.summary()

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```
